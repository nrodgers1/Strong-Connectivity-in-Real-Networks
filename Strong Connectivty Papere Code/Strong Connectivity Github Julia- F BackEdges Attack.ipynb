```julia
using LinearAlgebra, IterativeSolvers, Statistics

function get_levels(A)
    
    A= convert(Array{Float64}, A)
    
    w_in = vec(sum(A, dims=1))
   
    w_out= vec(sum(A, dims=2))
    
   
    
    u = w_in + w_out 
    
    v= w_in - w_out 
   
    Lambda=  diagm(u)-  A - transpose(A)
    
   
    Lambda[1,1]= 0 
    
    #h= bicgstabl(Lambda, v, 10)
    h=  lsmr(Lambda, v)   
       
    h= h .- minimum(h)
    
    return h
end
```

```julia
h = get_levels(A)
```

```julia
function calc_troph_incoh(A)
    
    F=0

    for (x,y,v) in zip(findnz(A)...)
  
    
        F = F + v*(h[y] - h[x] -1 )^2
        
    end
    
    F= F/sum(A)
    
end
    
    
    
    
    
```

```julia
using Random, Plots

#g = loadgraph("Ythan96.dat", "SimpleDiGraph", EdgeListFormat())
#g = loadgraph("large_n1000_500_graph_w_extra_500_path_len_0.edges", "SimpleDiGraph", EdgeListFormat())
#g = loadgraph("net_celegans_neural.dat", "SimpleDiGraph", EdgeListFormat())
#g = loadgraph("net_moreno_highschool.dat", "SimpleDiGraph", EdgeListFormat())
g= erdos_renyi(1000, 3*1000, is_directed=true)
#g = loadgraph("wiki-Vote.txt", "SimpleDiGraph", EdgeListFormat())
A = (adjacency_matrix(g))

h = get_levels(A)

F= calc_troph_incoh(A)
g1= copy(g)
g3 =copy(g)
g2=copy(g)

g4= copy(g)

strong_list = zeros(0)

strong_list_rand = zeros(0) 

strong_list_imb = zeros(0)

w_in = vec(sum(A, dims=1))
   
w_out= vec(sum(A, dims=2))



using LinearAlgebra
using Arpack

N=size(A)[1]
println(N)
diffs= zeros(0)
edge_rem= 0

x_list= zeros(0)
y_list =zeros(0)

x_list_imb= zeros(0)
y_list_imb =zeros(0)

diffs_imb= zeros(0)

for (x,y,v) in zip(findnz(A)...)
  
    
        d= (h[y] - h[x] )
        append!(diffs,d)
        append!(x_list,x)
        append!(y_list,y)
        append!(x_list_imb,x)
        append!(y_list_imb,y)
        append!(diffs_imb, (w_in[y] - w_out[y])- (w_in[x] - w_out[x]) )
    end


L= sum(A)

j = (sortperm(diffs))

p = (sortperm(diffs_imb))

x_list= floor.(Int,x_list[j])

y_list=  floor.(Int,y_list[j])

x_list_imb= floor.(Int,x_list_imb[p])

y_list_imb=  floor.(Int,y_list_imb[p])


rand_j= shuffle(j)

x_list_rand= (x_list[rand_j])

y_list_rand= (y_list[rand_j])

println(diffs[j])

weak_list= zeros(0)
weak_rand=zeros(0)

weak_list_imb =zeros(0)

L_list=zeros(0)
for i in 1:L
    #println(i)
    strong = strongly_connected_components(g1)   
    append!(strong_list, maximum(size.(strong))[1]/N)
    
    weak = weakly_connected_components(g1)
    append!(weak_list, maximum(size.(weak))[1]/N)
    rem_edge!(g1, x_list[i],y_list[i])
    
     strong_rand = strongly_connected_components(g2)   
    append!(strong_list_rand, maximum(size.(strong_rand))[1]/N)
    weak_2 = weakly_connected_components(g2)
    append!(weak_rand, maximum(size.(weak_2))[1]/N)
    
    rem_edge!(g2, x_list_rand[i],y_list_rand[i])
    
    
    strong_imb = strongly_connected_components(g4)   
    append!(strong_list_imb, maximum(size.(strong_imb))[1]/N)
    
    weak_imb = weakly_connected_components(g4)
    append!(weak_list_imb, maximum(size.(weak_imb))[1]/N)
    rem_edge!(g4, x_list_imb[i],y_list_imb[i])
    
    
    
    
    
    append!(L_list, i)
end


L_list= L_list/L


```
